```python
def random_plot():
    fig = plt.figure(figsize=(18, 15))
    cols = 3
    rows = 3
    for i in range(1, cols*rows +1):
        
        random_img = all_images[np.random.randint(0,len(all_images))]
        
        # read image
        img = mpimg.imread(data_path+ random_img)
        # get random image label
        img_lbl = list(all_labels[all_labels.file_name==random_img].category)[0]
        
        # add details in subplot
        fig.add_subplot(rows, cols, i)
        plt.title(img_lbl)
        plt.imshow(img)
        
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
```

### Exploring NLP aspect of the data

#### Creating Global word Vectorr

```python
embeddings_index = {}

# reading global vector corpus - takes a while to process
f = open('../input/global_vec/glove.840B.300d.txt', encoding='utf-8')

# iterating over each line and storing the vector of each word in 
for line in tqdm_notebook(f):
    try:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
    except ValueError:
        pass
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

#### creating word embeddings

```python
all_label_nlp = pd.read_csv(label_path+"all_label_nlp.csv")
```

```python
doc = u"Apple is looking at buying U.K. startup for $1 billion"
```

```python
def text_parser(doc):
    '''
    A text parser to clean the sentences
    '''
    # converting to lower case
    doc = doc.lower()
    
    # regex to remove special characters
    doc = re.sub("[!@#$%^&*()_\\-+={}[\]'<>?:;,/~`]", " ", doc)
    
    # removing stopwords (cause they are redundant)
    doc = remove_stopwords(doc)
    
    # replacing . with empty spaces
    doc = re.sub("[.]", "", doc)
    
    # tokenize into words
    doc = list(tokenize(doc))

    return doc
```

```python
text_parser(doc)
```

```python
# this function creates a normalized vector for the whole sentence
def sent2vec(s):
    M = []
    for w in s:
        try:
            M.append(embeddings_index[w])
        except:
            continue
    M = np.array(M)
    v = M.sum(axis=0)
    if type(v) != np.ndarray:
        return np.zeros(300)
    return v / np.sqrt((v ** 2).sum())
```

```python
x_test_nlp = np.array(all_label_nlp.meta.values[:no_of_test_points])
x_train_nlp = np.array(all_label_nlp.meta.values[no_of_test_points:])
```

```python
x_train_nlp.shape
```

```python
# cleaning sentences using our custom `text_parser` function
x_train_nlp = list(map(text_parser, x_train_nlp))
x_test_nlp = list(map(text_parser, x_test_nlp))
```

```python
x_test_nlp[:3]
```

```python
# generating word gobal vectors
x_train_globe = list(map(sent2vec, x_train_nlp))
x_test_globe = list(map(sent2vec, x_test_nlp))
```

```python
x_train_globe = np.array(x_train_globe)
x_test_globe = np.array(x_test_globe)
```

```python
pd.DataFrame(x_train_globe).to_csv('../input/global_vec/x_train_globe.csv', index=False)
pd.DataFrame(x_test_globe).to_csv('../input/global_vec/x_test_globe.csv', index=False)
```
